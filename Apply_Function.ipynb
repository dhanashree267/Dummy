```python
def get_actors_list(x):
    x = x.replace("u'", '')
    actors = re.sub('[^a-zA-Z,]+', ' ', x)
    actors_list = actors.split(',')
    actors_list = list(map(lambda x: x.strip(), actors_list))
    
    return actors_list

movies_df['actors_list'].apply(get_actors_list)
```

## Map Function

```python
For mapping values from dict or series with existing values in dataframe
majorly for creating new features
```

```python
movies_df1 = movies_df.head(10)
movies_df1
```

```python
movies_df1['genre'].replace({'Crime':'CRIME'})
```

```python
movies_df1['genre'].map({'Crime':'CRIME'})
```

```python
movies_df1['genre'].map({'Crime':'CRIME','Drama':'DRAMA'})
```

```python
dict1 = {'Player':['Rohit','Shubhman','Virat','Shreyas','KL'],
        'Jersy No': [45,77,18,96,1]}

df3 = pd.DataFrame(dict1)
df3
```

```python
scores = {'Rohit':40,'Virat':101, 'Shreyas':77}
df3['Score'] = df3['Player'].map(scores)
df3
```

```python
s1 = pd.Series(scores)
s1
```

```python
df3['Score2'] = df3['Player'].map(s1)
df3
```
